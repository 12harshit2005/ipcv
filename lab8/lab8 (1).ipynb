```python
import cv2
import numpy as np
```

```python
img = cv2.imread('/content/green image.jfif')
```

```python
gray = cv2.cvtColor(img, cv2.COLOR_BGR2GRAY)
```

```python
hsv = cv2.cvtColor(img, cv2.COLOR_BGR2HSV)
```

```python
h,s,v = cv2.split(hsv)
```

```python
print(hsv)
```

```python
lower_h = 35
upper_h = 85
```

```python
mask =cv2.inRange(h, lower_h, upper_h)
```

```python
result = cv2.bitwise_and(img, img, mask=mask)
```

```python
print(mask)
```

```python
cv2.imwrite('mask.jpg',result)
```
