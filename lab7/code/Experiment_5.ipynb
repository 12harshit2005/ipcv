```python
import cv2
import numpy as np
```

```python
img =cv2.imread('/content/image.jfif')
```

```python
rows, cols= img.shape[:2]
```

```python
M_trans = np.float32([[1,0,50],[0,1,50]])
```

```python
translated = cv2.warpAffine(img, M_trans, (cols,rows))
```

```python
cv2.imwrite("translation.jpg",translated)
```

```python
scaled = cv2.resize(img, None, fx=0.5, fy=0.5)
```

```python
cv2.imwrite("scaled.jpg",scaled)
```

```python
M_rot = cv2.getRotationMatrix2D((cols/2,rows/2),60,1)
```

```python
rotated = cv2.warpAffine(img,M_rot,(cols,rows))
```

```python
cv2.imwrite('rotated.jpg',rotated)
```
